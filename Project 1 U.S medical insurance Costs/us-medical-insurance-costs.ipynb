# U.S. Medical Insurance Costs

```python
import csv 

#Declare empty list for 6 columns
list_of_age = []
list_of_sex = []
list_of_bmi = []
list_of_children =[]
list_of_smoker = []
list_of_region = []
list_of_charges = []

#Define function to read the csv file
def load_list_data(lst, csv_file, column_name):
    with open(csv_file) as insurance_data:
        insurance_dict = csv.DictReader(insurance_data)
        for row in insurance_dict:
            lst.append(row[column_name])
    return lst
```

```python
#Read csv file to the empty lists
load_list_data(list_of_age, 'insurance.csv', 'age')
load_list_data(list_of_sex, 'insurance.csv', 'sex')
load_list_data(list_of_bmi, 'insurance.csv', 'bmi')
load_list_data(list_of_children, 'insurance.csv', 'children')
load_list_data(list_of_smoker, 'insurance.csv', 'smoker')
load_list_data(list_of_region, 'insurance.csv', 'region')
load_list_data(list_of_charges, 'insurance.csv', 'charges')
```

```python
#def class Patient_info
class PatientsInfo:
    def __init__(self,age,sex,bmi,children,smoker,region,charges):
        self.age = age                            #self.age is the list_of_age
        self.sex = sex                            #self.age is the list_of_sex
        self.bmi = bmi                            #self.age is the list_of_bmi
        self.children = children                  #self.age is the list_of_children
        self.smoker = smoker                      #self.age is the list_of_smoker
        self.region = region                      #self.age is the list_of_region
        self.charges = charges                    #self.age is the list_of_charges

#define a method that calculates the average ages of the patients in insurance.csv
    def analyze_average_ages(self):
        total_age = 0
        for age in self.age:
            total_age += int(age)
        average_age = round((total_age/len(self.age)),2)
        print("The average age of the patients is {}.".format(average_age))

#define a method that calculates the number of males and females in insurance.csv
    def analyze_sexes(self):
        num_male = 0
        num_female = 0
        for sex in self.sex:
            if sex == 'male':
                num_male += 1
            else:
                num_female += 1 
        print("The number of male patients is {}.".format(num_male))
        print("The number of female patients is {}.".format(num_female))
        
#define a method that find each unique region patients are from
    def unique_regions(self):
        unique_regions_list = []
        for region in self.region:
            if region not in unique_regions_list:
                unique_regions_list.append(region)
                
        return unique_regions_list

#define a method that find average yearly medical charges for patients in insurance.csv
    def average_charges(self):
        total_charges = 0
        for charge in self.charges:
            total_charges += float(charge)
        average_charge = round((total_charges / len(self.charges)),2)
        print("The average yearly medical charges for patients is {} dollars.".format(average_charge))

#define a method that creat a dictionary with all patients information
    def create_dictionary(self):    
        insurance_dict = {"age": self.age, "sex": self.sex, "bmi": self.bmi, "children": self.children, "smoker": self.smoker, "region": self.region, "charges": self.charges}
        return insurance_dict    

#define a method that analyze where a majorith of the individuals are from 
    def majority_come_from(self):
        maj_come_from = self.region[0]
        for region in self.region:
            if self.region.count(region) > self.region.count(maj_come_from):
                maj_come_from = region
            
        print("A majority of the individuals are from {}".format(maj_come_from))

#Define a method to find the charge different between smoker and non-smoker
    def difference_costs_of_smokervsnonsmo(self):
        for out_index in range(len(self.bmi)):
            for in_index in range(out_index + 1,len(self.bmi)):
                if (self.bmi[in_index]) == (self.bmi[out_index]):
                    if (self.age[out_index] == self.age[in_index]) and (self.sex[out_index] == self.sex[in_index]) and (self.children[out_index] == self.children[in_index]) and (self.region[out_index] == self.region[in_index]) and (self.smoker[out_index] != self.smoker[in_index]):
                        charges_difference = abs(float(self.charges[in_index])-float(self.charges[out_index]))
                        #print(out_index+2)
                        #print(in_index+2)
                        return print("The different costs between smokers vs. non-smokers is {}".format(charges_difference))
        return print("Sorry we can't find the charges difference")
    
#Define a method to figure out the average age is for someone who has at least one child
    def average_age_of_having_child(self):
        total_age_of_having_child = 0
        count_of_having_child = 0
        for index in range(len(self.children)):
            if int(self.children[index]) > 0:
                #print(self.age[index])
                total_age_of_having_child += int(self.age[index])
                count_of_having_child += 1
        average_age_having_child = round((total_age_of_having_child / count_of_having_child), 2)
        print("The average age is for someone who has at least one child is {} years old.".format(average_age_having_child))

#Define a method to find number of patients between 20 to 25 having at least one child
    def age_between2025_having_child(self):
        count_of_having_child = 0
        for index in range(len(self.children)):
            if int(self.children[index]) > 0 and (20 < int(self.age[index]) <= 25):
                #print(list_of_age[index])
                #total_age_of_having_child += int(list_of_age[index])
                count_of_having_child += 1
    #average_age_having_child = total_age_of_having_child / count_of_having_child
        print("{} of patients between 20 to 25 years old having at least one child.".format(count_of_having_child))

#Define a method that creat a detailed dictory of each Patient
    def created_detailed_dict(self):
        detailed_patientinfo_dict={}
        for patient_num in range(1,len(self.age)+1):
            detailed_patientinfo_dict["Patient No.{}".format(patient_num)] = {"age": int(self.age[patient_num-1]), "sex": self.sex[patient_num-1], "bmi": float(self.bmi[patient_num-1]), "children": int(self.children[patient_num-1]), "smoker": self.smoker[patient_num-1], "region": self.region[patient_num-1], "charges": float(self.charges[patient_num-1])}
            
        return detailed_patientinfo_dict
```

```python
#Define the object patient_intfo as class PatientsInfo
patient_info = PatientsInfo(list_of_age, list_of_sex, list_of_bmi, list_of_children, list_of_smoker, list_of_region, list_of_charges)       

#Analyze the average ages
patient_info.analyze_average_ages()

#calculates the number of males and females in insurance.csv
patient_info.analyze_sexes()

#find each unique region patients are from
print(patient_info.unique_regions())

#find average yearly medical charges for patients in insurance.csv
patient_info.average_charges()

#create dictionary with all patients information
#print(patient_info.create_dictionary())

#Analyze where a majority of the individuals are from.
patient_info.majority_come_from()

#Look at the different costs between smokers vs. non-smokers.  Or find the same paramethers but only different in the smoker
patient_info.difference_costs_of_smokervsnonsmo()

#Figure out what the average age is for someone who has at least one child in this dataset.
patient_info.average_age_of_having_child()

#Patients between 20 to 25 having at least one child
patient_info.age_between2025_having_child()

print(patient_info.created_detailed_dict())
```
